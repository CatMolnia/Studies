```python
import pandas as pd
```

```python
pd.read_csv("pokemon.csv")
```

```python
pokemon = pd.read_csv("pokemon.csv", index_col="Pokemon")["Type"]
pokemon
```

```python
pd.read_csv("google_stocks.csv", parse_dates=["Date"]).head()
```

```python
google = pd.read_csv("google_stocks.csv",
                     parse_dates=["Date"],
                     index_col="Date"
)["Close"]
google
```

```python
pd.read_csv("revolutionary_war.csv",
            index_col="Start Date",
            parse_dates=["Start Date"]
).tail()
```

```python
pd.read_csv(
    "revolutionary_war.csv",
    index_col="Start Date",
    parse_dates=["Start Date"],
    usecols=["State", "Start Date"]
).tail()
```

```python
battles = pd.read_csv(
    "revolutionary_war.csv",
    index_col="Start Date",
    parse_dates=["Start Date"],
    usecols=["State", "Start Date"]
)["State"].tail()
```
