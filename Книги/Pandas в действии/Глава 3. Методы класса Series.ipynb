"""---3.4 ПОДСЧЕТ КОЛИЧЕСТВА ЗНАЧЕНИЙ С ПОМОЩЬЮ МЕТОДА VALUE_COUNTS---"""

```python
pokemon.value_counts()
```

```python
pokemon.value_counts(ascending=True)
```

```python
pokemon.value_counts(normalize=True).head() # выводим соотношение количества частного к общему
```

```python
pokemon.value_counts(normalize=True).head() * 100 # то же самое в %
```

```python
(pokemon.value_counts(normalize=True) * 100).round(2) # округление до 2х знаков
```

```python
google.value_counts()
```

```python
google.max()
```

```python
google.max() - google.min()
```

```python
buckets = [0, 200, 400, 600, 800, 1000, 1200, 1400] # задаем интервал вхождений (корзины)
google.value_counts(bins = buckets) # считаем количество вхождений значений акций в этих интервалах ("(" - не включая; "]" - включая)
```

```python
google.value_counts(bins = buckets).sort_index()
```

```python
google.value_counts(bins = buckets, sort = False) # аналогично предыдущему коду
```

```python
google.value_counts(bins=6, sort=False).sort_index(ascending=True)
```

"""---3.5 ВЫЗОВ ФУНКЦИИ ДЛЯ КАЖДОГО ИЗ ЗНАЧЕНИЙ ОБЪЕКТА SERIES С ПОМОЩЬЮ МЕТОДА APPLY---"""

```python
funcs = [len, max, min]
```

```python
for current_func in funcs:
  print(current_func(google))
```

```python
google.apply(round) # округляем значения до целого числа
```

```python
def sindle_or_multi(x): # x - переменная (тип нашего покемона)
  if "/" in x:
    return "Multi"
  return "Single"
```

```python
pokemon.apply(sindle_or_multi)
```

```python
pokemon.apply(sindle_or_multi).value_counts()
```

```python
pd.read_csv("revolutionary_war.csv").head()
```

```python
days_of_war = pd.read_csv(
  "revolutionary_war.csv",
  usecols=["Start Date", "Battle"],
  parse_dates = ["Start Date"],
  index_col="Battle"
)["Start Date"]

days_of_war.head()
```

```python
def day_of_week(date):
  return date.strftime("%A")
```

```python
days_of_war.dropna().apply(day_of_week)
```

```python
days_of_war.dropna().apply(day_of_week).value_counts()
```
