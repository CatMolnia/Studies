```python
pd.read_csv(
    "revolutionary_war.csv",
    index_col="Start Date",
    parse_dates=["Start Date"],
    usecols=["State", "Start Date"]
).tail()
```

```python
battles = pd.read_csv(
    "revolutionary_war.csv",
    index_col="Start Date", # задаем индекс
    parse_dates=["Start Date"], # преобразоваваем столбец "Start Date" в тип даты (datetime)
    usecols=["State", "Start Date"] # выбираем только нужные столбцы при чтении файла
)["State"]
```

"""---3.2 СОРТИРОВКА ОБЪЕКТОВ SERIES---"""

"""---3.2.1 Сортировка значений с помощью метода sort_values---"""

```python
google.sort_values()
```

```python
pokemon.sort_values()
```

```python
pd.Series(data=["Adam", "adam", "Ben"]).sort_values() # сортируем в порядке возрастания (учитывая регистр)
```

```python
google.sort_values(ascending=False).head()
```

```python
pokemon.sort_values(ascending=False).head()
```

```python
battles.sort_values()
```

```python
battles.sort_values(na_position="last") # явно располагаем nan в конце списка
```

```python
battles.sort_values(na_position="first") # явно располагаем nan в начале списка
```

```python
battles.dropna().sort_values() # удаляем из списка пустые значения (не по index)
```

"""---3.2.2 Сортировка по индексу с помощью метода sort_index---"""

```python
pokemon.sort_index()
```

```python
battles.sort_index()
```

```python
battles.sort_index(na_position="first").head()
```

```python
battles.sort_index(ascending=False).head()
```

"""---3.2.3 Получение минимального и максимального значений с помощью методов nsmallest и nlargest---"""

```python
google.sort_values(ascending=False).head()
```

```python
google.nlargest() # возвращаем максимальные значения из списка (по умолчанию n = 5)
```

```python
google.nsmallest() # возвращаем минимальные значения из списка (по умолчанию n = 5)
```

"""---3.4 ПОДСЧЕТ КОЛИЧЕСТВА ЗНАЧЕНИЙ С ПОМОЩЬЮ МЕТОДА VALUE_COUNTS---"""

```python
pokemon.head()
```

```python
pokemon.value_counts()
```

```python
len(pokemon.value_counts()) # длина (количество) уникальных значений объекта Series
```

```python
pokemon.nunique() # аналогия
```

```python
pokemon.value_counts(ascending=True)
```

```python
pokemon.value_counts(normalize=True).head() # выводим соотношение количества частного к общему
```

```python
pokemon.value_counts(normalize=True).head() * 100 # то же самое в %
```

```python
(pokemon.value_counts(normalize=True) * 100).round(2) # округление до 2х знаков
```

```python
google.value_counts()
```

```python
google.max()
```

```python
google.min()
```

```python
google.max() - google.min()
```

```python
buckets = [0, 200, 400, 600, 800, 1000, 1200, 1400] # задаем интервал вхождений (корзины)
google.value_counts(bins = buckets) # считаем количество вхождений значений акций в этих интервалах ("(" - не включая; "]" - включая)
```

```python
google.value_counts(bins = buckets).sort_index()
```

```python
google.value_counts(bins = buckets, sort = False) # аналогично предыдущему коду
```
