"""---2.1.3. Пользовательские индексы для Series---"""

```python
import pandas as pd
import numpy as np
```

```python
ice_cream_flavors = [
  "Chocolate",
  "Vanilla",
  "Strawberry",
  "Rum Raisin",
]

pd.Series(ice_cream_flavors) # создание объекта Series
```

```python
days_of_week = ("Monday", "Wednesday", "Friday", "Saturday")
```

```python
pd.Series(ice_cream_flavors, days_of_week)
pd.Series(data = ice_cream_flavors, index = days_of_week)
```

```python
bunch_of_bools = [True, False, False]
pd.Series(bunch_of_bools)
```

```python
stock_prices = [985.32, 950.44]
time_of_day = ["Open", "Close"]
pd.Series(data = stock_prices, index = time_of_day)
```

```python
lucky_numbers = [4, 8, 15, 16, 23, 42]
pd.Series(lucky_numbers)
```

```python
lucky_numbers = [4, 8, 15, 16, 23, 42]
pd.Series(lucky_numbers, dtype = "float") # принудительно меняем формат передаваемых данных
```

"""---2.5.1. Статистические операции---"""

```python
numbers = pd.Series([1, 2, 3, np.nan, 4, 5])
numbers
```

```python
numbers.count() # подсчет непустых значений
```

```python
numbers.sum() # сумма непустых значений
```

```python
numbers.sum(skipna = False) # определение суммы значений с пропущенными значениями
```

```python
numbers.sum(min_count = 3) # определение суммы значений с минимальным количеством непустых значений
```

```python
numbers.product() # перемнажаем все значения Series
```

```python
numbers.product(min_count=3)
```

```python
numbers.cumsum() # нарастающая сумма
```

```python
numbers.cumsum(skipna = False) # рассчет нарастающей суммы до первого пустого знаяниея
```

```python
numbers.pct_change() # подтягивает вместо nan последнее (предыдущее nan) допустимое значение (также можно через 'pad' или 'ffill')
```

```python
numbers.pct_change(fill_method="bfill") # подтягивает вместо nan следуюшее (первое после nan) допустимое значение (также можно через 'bfill' или 'backfill')
```

```python
numbers.mean() # вывод среднего значения (nan не учитывается)
```

```python
numbers.median() # вывод значения, распопложенного посередине в отсортированном списке значений Series (nan не учитывается)
```

```python
numbers.std() # вывод стандартного отклонения (мера отклонения данных от среднего значения)
```

```python
numbers.max() # извлекаем максимальное значение
```

```python
numbers.min() # извлекаем минимальное значение
```

```python
animals = pd.Series(["koala", "aardvark", "zebra"])
animals
```

```python
animals.max() # ближайшая к концу алфавита буква
```

```python
numbers.describe() # вся сводная информация об объекте Series
```

```python
numbers.sample(3) # извлекаем случайный набор значений из объекта Series
```

"""---2.5.2. Арифметические операции---"""

```python
s1 = pd.Series(data = [5, np.nan, 15], index = ["A", "B", "C"])
s1
```

```python
s1 + 3 # операция сложения
```

```python
s1.add(3) # аналог операция сложения
```

```python
s1 - 5 # операция вычитания
s1.sub(5) # аналог
s1.subtract(5) # аналог
```

```python
s1 * 2 # операция умножения
s1.mul(2) # аналог
s1.multiply(2) # аналог
```

```python
s1 / 2 # операция деления
s1.div(2) # аналог
s1.divide(2) # аналог
```

```python
s1 // 4 # целочисленное деление (округление вниз)
s1.floordiv(4) # аналог
```

```python
s1 % 3
s1.mod(3)
```

"""---2.5.3. Транслирование---"""

```python
s1 = pd.Series([1, 2, 3], index = ["A", "B", "C"])
s2 = pd.Series([4, 5, 6], index = ["A", "B", "C"])
s1 + s2
```

```python
s1 = pd.Series(data = [3, 6, np.nan, 12])
s2 = pd.Series(data = [2, 6, np.nan, 12])
s1 == s2 # проверяем два объекта на равенство
```

```python
s1 != s2 # проверяем два объекта на неравенство
s1.ne(s2) # аналог
```

```python
s1 = pd.Series(
  data = [5, 10, 15], index = ["A", "B", "C"]
)
s2 = pd.Series(
  data = [4, 8, 12, 14], index = ["B", "C", "D", "E"]
)
s1 + s2
```
