"""---2.1.3. Пользовательские индексы для Series---"""

```python
import pandas as pd
import numpy as np
```

```python
ice_cream_flavors = [
  "Chocolate",
  "Vanilla",
  "Strawberry",
  "Rum Raisin",
]

pd.Series(ice_cream_flavors) # создание объекта Series
```

```python
days_of_week = ("Monday", "Wednesday", "Friday", "Saturday")
```

```python
pd.Series(ice_cream_flavors, days_of_week)
pd.Series(data = ice_cream_flavors, index = days_of_week)
```

```python
bunch_of_bools = [True, False, False]
pd.Series(bunch_of_bools)
```

```python
stock_prices = [985.32, 950.44]
time_of_day = ["Open", "Close"]
pd.Series(data = stock_prices, index = time_of_day)
```

```python
lucky_numbers = [4, 8, 15, 16, 23, 42]
pd.Series(lucky_numbers)
```

```python
lucky_numbers = [4, 8, 15, 16, 23, 42]
pd.Series(lucky_numbers, dtype = "float") # принудительно меняем формат передаваемых данных
```

"""---2.2 СОЗДАНИЕ ОБЪЕКТОВ SERIES НА ОСНОВЕ ОБЪЕКТА ЯЗЫКА PYTHON---"""

```python
# ассоциативный массив (словарь) (ключ: значение)
calorie_info = {
  "Cereal": 125,
  "Chocolate Bar": 406,
  "Ice Cream Sundae": 342,
}

diet = pd.Series(calorie_info)

diet
```

```python
# кортежи
pd.Series(data = ("Red", "Green", "Blue")) # кортеж (неизменяемый список)
```

```python
# кортежи в списке
rgb_colors = [(120, 41, 26), (196, 165, 45)] # кортежи, заключенные в список
pd.Series(data = rgb_colors)
```

```python
# множество
my_set = {"Ricky", "Bobby"}
pd.Series(list(my_set)) # преобразуем множество в список (иначе будет ошибка)
```

```python
# объект ndarray библиотеки NymPy
random_data = np.random.randint(1, 101, 10)
random_data
```

```python
pd.Series(random_data) # передаем в Series объект ndarray
```

"""---2.3 АТРИБУТЫ SERIES---"""

```python
diet = {
    'Cereal': 125,
    'Chocolate Bar': 406,
    'Ice Cream Sundae': 342,
}
```

```python
np.array(diet.values()) # обращаемся к объекту ndarray, в котором храняться значения
```

```python
pd.Series(diet).index # обращаемся к Series для определения индекса
```

```python
pd.Series(diet).dtype # обращаемся к Series для определения типа данных
```

```python
pd.Series(diet).size # обращаемся к Series для определения количества элементов
```

```python
pd.Series(diet).shape # обращаемся к Series для определения формы данных
```

```python
pd.Series(diet).is_unique # обращаемся к Series для определения уникальности данных
```

```python
pd.Series(data=[3, 2, 1]).is_monotonic_decreasing # обращаемся к Series для определения монотонности данных (убывание)
```

```python
pd.Series(data=[1, 3, 2]).is_monotonic_increasing # обращаемся к Series для определения монотонности данных (возрастание)
```

```python
numbers = pd.Series([1, 2, 3, np.nan, 4, 5])
numbers
```

```python
numbers.sum() # сумма непустых значений
```
