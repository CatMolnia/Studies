"""---4.1.1 Создание объекта DataFrame на основе ассоциативного массива---"""

```python
import pandas as pd
import numpy as np
```

```python
city_data = {
    "City": ["New York", "Paris", "Barcelona", "Rome"],
    "Country": ["United States", "France", "Spain", "Italy"],
    "Population": [8600000, 2141000, 5515000, 2873000]
}

cities = pd.DataFrame(city_data)
cities
```

```python
cities.T # транспонировали таблицу
```

"""---4.1.2 Создание объекта DataFrame на основе ndarray библиотеки NumPy---"""

```python
random_data = np.random.randint(1, 101, [3, 5])
random_data
```

```python
pd.DataFrame(data=random_data)
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
temperatures = pd.DataFrame(random_data,
                            index=row_labels)
temperatures
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
column_labels = (
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday"
)
temperatures = pd.DataFrame(random_data,
                            index=row_labels,
                            columns=column_labels)
temperatures
```

```python
row_labels = ["Morning", "Afternoon", "Morning"]
column_labels = (
    "Monday",
    "Tuesday",
    "Wednesday",
    "Tuesday",
    "Friday"
)
temperatures = pd.DataFrame(random_data,
                            index=row_labels,
                            columns=column_labels)
temperatures
```
